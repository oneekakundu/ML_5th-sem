```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import model_selection
from sklearn.svm import SVC
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import train_test_split
```

```python
#df= pd.read_csv("C:\\Users\\ONEEKA\Downloads\\penguins_binary.csv")
df = pd.read_csv("C:\\Users\\ONEEKA\\Downloads\\penguins.csv")
```

```python
from sklearn.preprocessing import LabelEncoder

LE = LabelEncoder()

# Create new columns for encoded values
df['Species_n'] = LE.fit_transform(df['species'])
df['Island_n']  = LE.fit_transform(df['island'])
```

```python
df['Gender_n'] = LE.fit_transform(df['Gender'])
```

```python
df
```

```python
df.drop(axis=1,columns='Gender',inplace=True)
df.drop(axis=1,columns='species',inplace=True)
df.drop(axis=1,columns='island',inplace=True)
```

```python
df
```

```python
X = df.drop('Gender_n', axis=1)
y = df['Gender_n']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20,random_state=5)
```

```python
# PROCESS 1
model = BaggingClassifier(SVC())
```

```python
kfold=model_selection.KFold(n_splits=10)
results = model_selection.cross_val_score(model,X_train, y_train, cv=kfold)
```

```python
print("Results:", results)
print("Mean Results:", results.mean())
```

```python
#PROCESS 2
model = BaggingClassifier(SVC())
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
