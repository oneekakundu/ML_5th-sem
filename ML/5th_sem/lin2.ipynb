```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error,r2_score
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("C:\\Users\\ONEEKA\\Downloads\\Boston.csv")
df.head()
```

```python
X=df[['lstat']]
Y=df['medv']
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
model=LinearRegression()
model.fit(xtrain,ytrain)
```

```python
pred=model.predict(xtest)
pred
```

```python
model.score(xtest,ytest)
```

```python
model.score(xtrain,ytrain)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.scatter(xtrain,ytrain)
plt.plot(xtest,pred,color='r')
```

```python
plt.scatter(xtest,ytest)
plt.plot(xtest,pred,color='r')
```

```python
print("RMSE: ",root_mean_squared_error(ytest,pred))
print("MSE: ",mean_squared_error(ytest,pred))
print("AMSE: ",mean_absolute_error(ytest,pred))
print("R^2: ",r2_score(ytest,pred))
```
