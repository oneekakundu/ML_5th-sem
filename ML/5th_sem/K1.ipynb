```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv("C:\\Users\\ONEEKA\\Downloads\\Iris.csv")
df
```

```python
df.describe()
df.info()
```

```python
df.head()
```

```python
df=df.drop(columns=['Species','Id'])
df
```

```python
model=MinMaxScaler()
df=model.fit_transform(df)
df
```

```python
df=pd.DataFrame(df,columns=['A','B','C','D'])
df
```

```python
plt.scatter(df['A'],df['B'])
```

```python
model=KMeans(n_clusters=2)
model.fit(df)
```

```python
pred=model.predict(df)
pred
```

```python
df['Cluster']=pred
df
```

```python
cent=model.cluster_centers_
cent
```

```python
plt.scatter(df['A'],df['B'])
```

```python
import seaborn as sns
sns.scatterplot(df,x=df['A'],y=df['B'],hue=model.labels_)
```

```python
k_rng=range(1,11)
Sse=[]

for k in k_rng:
    model=KMeans(n_clusters=k)
    model.fit(df[['A','B']])
    Sse.append(model.inertia_)

Sse
```
