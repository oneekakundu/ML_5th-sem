```python
import numpy as np
x = np.arange(0,100,1)
x
y = x*2
y
z = x**2
z
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
#automatically shows plots witout plt.show()
```

```python
fig = plt.figure()# Create a blank canvas (Figure)
axes = fig.add_axes([0,0,1,1]) # Add a plotting area (Axes) to the canvas
axes.plot(x,y,'r--') #Draw a line on that plot
axes.set_xlabel('x axis')
axes.set_ylabel('y axis')
axes.set_title('x,y graph')
plt.show()
```

```python
fig = plt.figure()
ax1= fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
ax1.plot(x,y,'r--')
ax1.set_xlabel('x axis')
ax1.set_ylabel('y axis')
ax1.set_title('ax1')
ax2.plot(x,y,'b--')
ax2.set_xlabel('x axis')
ax2.set_ylabel('y axis')
ax2.set_title('ax2')
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
ax1.plot(x,y,'r--')
ax1.set_xlabel('x axis')
ax1.set_ylabel('y axis')
ax1.set_title('ax1')
ax2.plot(x,y,'b--')
ax2.set_xlabel('x axis')
ax2.set_ylabel('y axis')
ax2.set_title('ZOOM')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
plt.show()
```

```python
fig,axes=plt.subplots(1,2)
axes[0].plot(x,y,'r--',lw=3)
axes[1].plot(x,z,'m-',lw=5)
plt.show()
```

```python
fig,axes=plt.subplots(1,2,figsize=(12,3))
axes[0].plot(x,y,'r--',lw=3)
axes[1].plot(x,z,'m--',lw=5)
plt.show()
```

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,2*x,label='2*x',color ='b',alpha=0.6,lw=3,marker='o',mfc='y',mec ='g')
ax.plot(x,z,label='x**2',color='m',lw=5,ls='--')
ax.legend(loc=0)
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_title('x,ygraph')
plt.show()
```

```python
#Creating a Boxplot for Two Columns
data = {"column1": [1, 2, 3, 4, 5], "column2": [2, 4, 6, 8, 10]}
df = pd.DataFrame(data)
df.boxplot(column=["column1", "column2"])#create boxplot
plt.title("Boxplot of Two Columns")
plt.xlabel("Columns")
plt.ylabel("Values")
plt.show()
```

```python
# Load the dataset #Custom Box Colors
data = {"column1": [1, 2, 3, 4, 5], "column2": [2, 4, 6, 8, 10]}
df = pd.DataFrame(data)
box_colors = ["b", "r"]
fig, ax = plt.subplots() # Create the boxplot with custom colors
bp = ax.boxplot([df["column1"], df["column2"]], patch_artist=True)
for patch, color in zip(bp['boxes'], box_colors): # Apply custom colors to each box
    patch.set_facecolor(color)
plt.title("Boxplot with Custom Box Colors")
plt.xlabel("Columns")
plt.ylabel("Values")
plt.show()
```

```python
df.boxplot(column=["column1", "column2"], vert=False) #Horizontal Boxplot
plt.title("Horizontal Boxplot")
plt.xlabel("Values")
plt.ylabel("Columns")
plt.show()
```

```python
#Show Outlier Points
data = {"column1": [1, 2, 3, 4, 20], "column2": [2, 4, 6, 8, 10]}
df = pd.DataFrame(data)
df.boxplot(column=["column1", "column2"], showfliers=True, flierprops=dict(marker="o", markerfacecolor="red", markersize=8))
plt.title("Boxplot with Outlier Points")
plt.xlabel("Columns")
plt.ylabel("Values")
plt.show()
```
